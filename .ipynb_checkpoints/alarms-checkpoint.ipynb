```python
from numpy import fft
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd
from scipy import signal
from scipy.io import wavfile
```

**Code pour le signal de l'hyperglycémie**

```python
A = 4
duree = 0.15
Fe = 22000
Te = 1 / Fe

frequencies = [1312, 1410, 1500, 1619, 1722]

t = np.arange(0, duree + Te, Te)

signals = []

for i in range(len(frequencies)):
    signal = A * np.sin(2 * np.pi * frequencies[i] * t)  
    signals.extend(signal)

signal = np.array(signals)

sd.play(signal, samplerate=Fe)
```

**Code pour le signal de l'hypoglycémie**

```python
A = 4
duree = 0.15
Fe = 22000
Te = 1 / Fe

frequencies = [1722, 1619, 1500, 1410, 1312]

t = np.arange(0, duree + Te, Te)

signals = []

for i in range(len(frequencies)):
    signal = A * np.sin(2 * np.pi * frequencies[i] * t)  
    signals.extend(signal)

signal = np.array(signals)

sd.play(signal, samplerate=Fe)
```

**Test**

```python
def filter_signal(s, Wn1, Wn2, type, Fs):
    Wn = [Wn1 / (Fs / 2), Wn2 / (Fs / 2)] 
    b, a = signal.butter(1, Wn, type) 
    
    s = signal.filtfilt(b, a, s)

    return s

def identify_signal(file):
    Fs, s = wavfile.read(file)  

    s = filter_signal(s, 1311, 1723, 'band', Fs)

    s = filter_signal(s, 1313, 1411, 'stop', Fs)

    s = filter_signal(s, 1413, 1499, 'stop', Fs)

    s = filter_signal(s, 1501, 1618, 'stop', Fs)

    s = filter_signal(s, 1620, 1721, 'stop', Fs)

    sd.play(s, Fs)
    
    f, t, Sxx = signal.spectrogram(s, Fs, window='hamming', nperseg=1024, noverlap=512)
    plt.pcolormesh(t, f, np.log10(Sxx))
    plt.ylabel('Fréquences')
    plt.xlabel('Temps')
    plt.show()

identify_signal('alarms/moustique_hyper_mono.wav')
```
